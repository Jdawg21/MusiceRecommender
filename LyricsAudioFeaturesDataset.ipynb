# Selecting required columns for recommendation

```python
song_data = song_df[['track_name', 'track_artist', 'lyrics',
      'track_album_name',
       'playlist_name',  'playlist_genre', 'playlist_subgenre',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
        'language']]
```

```python
#song_data[song_data['language'].isin(['en','es'])]
```

```python
#song_data = song_data[song_data['language'].isin(['en','es'])]
song_data = song_data[song_data['language'].isin(['en'])]
song_data.reset_index(inplace=True)
song_data.shape
```

```python
print("English makes up ", np.round((song_data.shape[0]/song_df.shape[0])*100,3), "% of the data")
```

```python
song_data.drop('index',axis=1,inplace=True)
song_data.head()
```

# Convert all str to lower case

```python
song_data.head(3)
```

```python
song_data['playlist_name'] = song_data['playlist_name'].str.lower()
song_data['track_name'] = song_data['track_name'].str.lower()
song_data['track_artist'] = song_data['track_artist'].str.lower()
song_data['track_album_name'] = song_data['track_album_name'].str.lower()
song_data['lyrics'] = song_data['lyrics'].str.lower()
song_data['playlist_genre'] = song_data['playlist_genre'].str.lower()
song_data['playlist_subgenre'] = song_data['playlist_subgenre'].str.lower()
```

```python
song_data.head()
```

# drop duplicate records

```python
song_data = song_data.drop_duplicates(subset=['track_name','track_artist'], keep='first')
song_data.reset_index(inplace=True)
song_data.drop('index',axis=1,inplace=True)
```

# check if the numeric features follow Gaussian distribution

```python
song_data.columns
```

```python
from scipy.stats import shapiro

stat,p =shapiro(song_data[['danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo']])

print("stat: ",stat,":","p :",p)
if p>0.05:
    print("gaussian distribution - fail to reject H0")
else:
    print("not gaussian distribution - reject H0")
```

# Standardise using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler


minmaxscaler = MinMaxScaler()
minmaxscaled = minmaxscaler.fit_transform(song_data[['danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo']])
songs_normalized = pd.DataFrame(minmaxscaled, columns=[['danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo']])
```

```python
songs_normalized
```

```python
song_df_normalized = song_data.copy()
song_df_normalized.head()

```

```python

song_df_normalized[['danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo']] = songs_normalized
#song_df_normalized.drop(['bpm','nrgy','dnce','dB','live','val','acous','spch','pop'],axis=1,inplace=True)
song_df_normalized.head()
```

```python
song_df_normalized.tail()
```

# create feature matrix

```python
song_features=song_df_normalized.set_index("track_name")
song_features.drop(['track_artist', 'lyrics', 'track_album_name',
       'playlist_name', 'playlist_genre', 'playlist_subgenre','language'],axis=1,inplace=True)
song_features.head()
```

```python
song_features.info()
```

```python
song_features.tail()
```

# Use cosine similarity based NN algo to calculate min distances and find similar songs

```python
from scipy.sparse import csr_matrix

song_features_csr = csr_matrix(song_features.values)
```

```python
song_features_csr 
```

```python
song_features.values
```

```python
from sklearn.neighbors import NearestNeighbors

model_nn = NearestNeighbors(metric='cosine',algorithm='brute')
model_nn.fit(song_features_csr)
```

```python
recommened_song_list = []
```

```python
temp = song_features.copy()
temp.reset_index(inplace=True)
songsearch = 'crazy little thing called love - remastered 2011'
songsearch = songsearch.lower()
song_index = temp.index[temp['track_name'] ==songsearch].tolist()[0]
print(song_index)

print(song_features.index[song_index])


distances,indices = model_nn.kneighbors(X = song_features.iloc[song_index,:].values.reshape(1,-1), n_neighbors=6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print("Recommendation for ",song_features.index[song_index],"are: ")
    else:
        print(i,": ",song_features.index[indices.flatten()[i]], "| distance= ",distances.flatten()[i])
        
for i in range(1,4): 
    recommened_song_list.append(song_features.index[indices.flatten()[i]])
```

# Improved Recommendation

```python
songgname = 'crazy little thing called love - remastered 2011'
songgname = songgname.lower()
songg = song_data[song_data['track_name'] == songgname]
artist = songg['track_artist'].unique().tolist()[0]
print("Artist / Band is: ",artist)


genre = songg['playlist_genre'].unique().tolist()[0]
print("Genre is: ",genre)

subgenre = songg['playlist_subgenre'].unique().tolist()[0]
print("Subgenre is: ",subgenre)
```

## recommending songs of same artist 

```python
artist_data = song_df_normalized[song_df_normalized['track_artist'] == artist]
artist_song_features=artist_data.set_index("track_name")
artist_song_features.drop(['track_artist', 'lyrics', 'track_album_name',
       'playlist_name', 'playlist_genre', 'playlist_subgenre','language'],axis=1,inplace=True)
#artist_song_features.head()
#from scipy.sparse import csr_matrix

artist_song_features_csr = csr_matrix(artist_song_features.values)

model_nn = NearestNeighbors(metric='cosine',algorithm='brute')
model_nn.fit(artist_song_features_csr)

temp = artist_song_features.copy()
temp.reset_index(inplace=True)
songsearch = 'crazy little thing called love - remastered 2011'
songsearch = songsearch.lower()
song_index = temp.index[temp['track_name'] ==songsearch].tolist()[0]
print(song_index)

print(artist_song_features.index[song_index])


distances,indices = model_nn.kneighbors(X = artist_song_features.iloc[song_index,:].values.reshape(1,-1), n_neighbors=6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print("Recommendation for ",artist_song_features.index[song_index],"are: ")
    else:
        print(i,": ",artist_song_features.index[indices.flatten()[i]], "| distance= ",distances.flatten()[i])
        
for i in range(1,4): 
    recommened_song_list.append(artist_song_features.index[indices.flatten()[i]])
```

## recommending songs from same genre

```python
genre_data = song_df_normalized[song_df_normalized['playlist_genre'] == genre]
genre_song_features=genre_data.set_index("track_name")
genre_song_features.drop(['track_artist', 'lyrics', 'track_album_name',
       'playlist_name', 'playlist_genre', 'playlist_subgenre','language'],axis=1,inplace=True)
#genre_song_features.head()
#from scipy.sparse import csr_matrix

genre_song_features_csr = csr_matrix(genre_song_features.values)

model_nn = NearestNeighbors(metric='cosine',algorithm='brute')
model_nn.fit(genre_song_features_csr)

temp = genre_song_features.copy()
temp.reset_index(inplace=True)
songsearch = 'crazy little thing called love - remastered 2011'
songsearch = songsearch.lower()
song_index = temp.index[temp['track_name'] ==songsearch].tolist()[0]
print(song_index)

print(genre_song_features.index[song_index])


distances,indices = model_nn.kneighbors(X = genre_song_features.iloc[song_index,:].values.reshape(1,-1), n_neighbors=6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print("Recommendation for ",genre_song_features.index[song_index],"are: ")
    else:
        print(i,": ",genre_song_features.index[indices.flatten()[i]], "| distance= ",distances.flatten()[i])
        
for i in range(1,4): 
    recommened_song_list.append(genre_song_features.index[indices.flatten()[i]])
```

## recommending songs from same subgenre

```python
subgenre_data = song_df_normalized[song_df_normalized['playlist_subgenre'] == subgenre]
subgenre_song_features=subgenre_data.set_index("track_name")
subgenre_song_features.drop(['track_artist', 'lyrics', 'track_album_name',
       'playlist_name', 'playlist_genre', 'playlist_subgenre','language'],axis=1,inplace=True)
#subgenre_song_features.head()
#from scipy.sparse import csr_matrix

subgenre_song_features_csr = csr_matrix(subgenre_song_features.values)

model_nn = NearestNeighbors(metric='cosine',algorithm='brute')
model_nn.fit(subgenre_song_features_csr)

temp = subgenre_song_features.copy()
temp.reset_index(inplace=True)
songsearch = 'crazy little thing called love - remastered 2011'
songsearch = songsearch.lower()
song_index = temp.index[temp['track_name'] ==songsearch].tolist()[0]
print(song_index)

print(subgenre_song_features.index[song_index])


distances,indices = model_nn.kneighbors(X = subgenre_song_features.iloc[song_index,:].values.reshape(1,-1), n_neighbors=6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print("Recommendation for ",subgenre_song_features.index[song_index],"are: ")
    else:
        print(i,": ",subgenre_song_features.index[indices.flatten()[i]], "| distance= ",distances.flatten()[i])
        

for i in range(1,4): 
    recommened_song_list.append(subgenre_song_features.index[indices.flatten()[i]])
```
